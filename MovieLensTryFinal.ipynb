# Can we convert these binary columns in "movies" to one column so we can do group by statements?

```python
import pandas as pd
import numpy as np
```

```python
movie = pd.read_csv("movie2.csv")
movie.head()
```

```python
#getting column names
movie_genre = movie
```

```python
for i in movie_genre.loc[:,'Action':'Western'].columns:
    movie_genre[i] = movie_genre[i].replace(1,i)
movie_genre.head()
```

```python
movie_genre['genre'] = movie_genre.loc[:,'Action':'Western'].apply(set,axis=1)
movie_genre.head()
```

```python
movie_genre['genre'].apply(lambda x: x.remove(0))
movie_genre.head()
```

```python
movie_genre = movie_genre[['movie title','release date','genre']]
movie_genre.head()
```

```python
movie_genre = movie_genre['genre'].apply(lambda x: list(x)).explode()
movie_genre.head()
```

```python
genres = ['Action', 'Adventure',
       'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary', 'Drama',
       'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance',
       'Sci-Fi', 'Thriller', 'War', 'Western','genre']
genres
```

```python
movie.drop(genres,axis=1).merge(movie_genre,left_index=True,right_index=True)
```
