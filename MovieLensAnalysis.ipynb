```python
#finding movies which have more than one genre
#gather the unique movie ids and titles
movie_list = movies[['movie id', 'movie title']]
movie_list['genre_count'] = movies[['movie title', 'release date', 'Action', 'Adventure',
       'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary', 'Drama',
       'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance',
       'Sci-Fi', 'Thriller', 'War', 'Western']].sum(axis=1)
```

```python
movie_list[movie_list['genre_count']>1]
```

```python
#find high rated movies
#join ratings and movies
movie_ratings = movies.merge(ratings, on='movie id',how='inner')
movie_ratings.head()
```

```python
movie_ratings.shape
```

```python
avg_ratings = movie_ratings.groupby('movie title')['rating'].mean()
avg_ratings.sort_values(ascending=False).head(25)
```

```python
#find gender distribution across genre
user_movie_ratings = movie_ratings.merge(users,on='user id',how='inner')
user_movie_ratings
```

```python
gender_ratings = user_movie_ratings.groupby('gender').sum().loc[:,'Action':'Western']
gender_ratings
```
