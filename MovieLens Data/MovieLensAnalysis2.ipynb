# What are the average ratings by genre?

```python
import pandas as pd
import numpy as np
```

```python
# Reading datasets by using read_csv from pandas package
ratings = pd.read_csv("ratings.csv")
movie = pd.read_csv("movie.csv",index_col='movie id')
```

```python
#get ratings by genre
movie_ratings = movie.merge(ratings,on='movie id',how='inner')
movie_ratings.head()
```

```python
movie_ratings.columns
```

```python
movie_ratings.shape
```

```python
genre_movies = movie_ratings.loc[:,'Action':'Western']
genre_movies.head()
```

```python
genre_movies.shape
```

```python
genre_ratings = genre_movies.multiply(movie_ratings['rating'], axis="index")
genre_ratings.head()
```

```python
genre_ratings = genre_ratings.replace(0, np.NaN)
genre_ratings.mean()
```

```python
movie_ratings.to_csv('movie_ratings.csv')
```
