# Can we convert these binary columns in "movies" to one column so we can do group by statements?

```python
import pandas as pd
import numpy as np
```

```python
# Reading datasets by using read_csv from pandas package
movie = pd.read_csv("movie.csv")#,index_col='movie id')
```

```python
movie.head()
```

```python
#getting column names
movie_genre = movie
for i in movie_genre.loc[:,'Action':'Western'].columns:
    movie_genre[i] = movie_genre[i].replace(1,i)
movie_genre.replace(0,np.NaN)
movie_genre['genre'] = movie_genre.loc[:,'Action':'Western'].apply(set,axis=1)
movie_genre['genre'].apply(lambda x: x.remove(0))
movie_genre = movie_genre[['movie title', 'release date', 'genre']]
movie_genre = movie_genre['genre'].explode()
movie_genre.head()
```

```python
moviev2 = movie
for gn in moviev2.loc[:,'Action':'Western'].columns:
    moviev2[gn] = moviev2[gn].replace(1, gn)
moviev2['genre'] = moviev2.loc[:,'Action':'Western'].apply(set,axis=1)
moviev2['genre'].apply(lambda x: x.remove(0))
moviev2 = moviev2[['movie title','release date','genre']]
movie_genre = moviev2['genre'].explode()#apply(lambda x: ",".join(x).explode())
movie_genre.reset_index()
moviev2.head()
moviev2.drop(moviev2[['genre']],axis=1).merge(movie_genre,left_index=True,right_index=True)
```
